```python
def pascal(iterations):
    """
    Takes in the argument "iteration" and returns a triangle with that many iterations of random choice.
    """
    
    #create the points of the vertecies of a trianlge
    x = [0,1,1/2]
    y = [0, 0, np.sqrt(3)/2]
    
    #create emtpy lists
    x_list = [0]
    y_list = [0]
    

    #start point for this iteration
    x_temp = 0
    y_temp = 0
    
    #creates random list of len iteration from 1 to 3
    choice_list = np.random.randint(1,4,iterations)
    
    for char in range(iterations):
        #this chooses the index
        if choice_list[char] == 1:
            #averages the current point with the corresponding coordinates of the vertex to get the new point
            x_temp = (x_temp + 0)/2
            y_temp = (y_temp + 0)/2
            x_list.append(x_temp)
            y_list.append(y_temp)
            
        if choice_list[char] == 2:
            
            x_temp = (x_temp + 1)/2
            y_temp = (y_temp + 0)/2
            x_list.append(x_temp)
            y_list.append(y_temp)
            
        if choice_list[char] == 3:
            
            x_temp = (x_temp + (1/2))/2
            y_temp = (y_temp + (np.sqrt(3)/2))/2
            x_list.append(x_temp)
            y_list.append(y_temp)
        
        
        
    #plots the triangle  
    plt.plot(x,y, c="C0")
    plt.plot(x[::len(x)-1] ,y[::len(y)-1], c="C0") 
    plt.scatter(x_list, y_list, s=3, c="C1", alpha = .7)
        
```

```python
pascal(100000)
```
